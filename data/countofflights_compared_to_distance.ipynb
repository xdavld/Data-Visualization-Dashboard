```python
#Anzahl der Fl√ºge pro Airline im Vergleich zur geflogenen Distanz
import pandas as pd

df = pd.read_csv('./all_flights_with_lon_and_lat.csv')

airline_counts = df.groupby('AIRLINE').agg({'FLIGHT_NUMBER': 'count', 'DISTANCE': 'sum'})
airline_counts = airline_counts.sort_values('FLIGHT_NUMBER', ascending=True)

airline_counts.head(100)
```

```python
airline_counts.shape
airline_counts.to_csv('./airline_counts.csv')
```
