```python
import pandas as pd

df = pd.read_csv('./all_flights_with_lon_and_lat.csv')

mean_arrival_delay = df.groupby(['AIRLINE', 'MONTH'])['ARRIVAL_DELAY'].mean()
mean_arrival_delay = mean_arrival_delay.reset_index()

mean_arrival_delay.head(100)
```

```python
mean_arrival_delay.shape
mean_arrival_delay.to_csv('./mean_arrival_delay.csv', index=False)
```
