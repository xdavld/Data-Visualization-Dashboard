```python
import pandas as pd
df = pd.read_csv("../Documents/flights.csv", low_memory=False)

df = df[df["CANCELLED"] == 0]
df.drop(["TAXI_IN","WHEELS_ON","CANCELLED","YEAR","DIVERTED","WHEELS_OFF",'TAXI_OUT',"TAIL_NUMBER","CANCELLATION_REASON","AIR_SYSTEM_DELAY",'SECURITY_DELAY','AIRLINE_DELAY','LATE_AIRCRAFT_DELAY','WEATHER_DELAY'],axis=1,inplace=True)

df.dropna(axis=0,inplace=True)
```

```python
df.head()
```

```python
count = 0
for airport_name in location_daten.values:
    name = airport_name[0]
    lon = airport_name[-1]
    lat = airport_name[-2]
    count += 1
    print(count)
    df.loc[df[df['DESTINATION_AIRPORT'] == name].index, "DESTINATION_AIRPORT_LAN"] =  lat
    df.loc[df[df['DESTINATION_AIRPORT'] == name].index, "DESTINATION_AIRPORT_LON"] =  lon
    
```

```python
df.rename(columns={'DESTINATION_AIRPORT_LAN': 'DESTINATION_AIRPORT_LAT'}, inplace=True) 
```
